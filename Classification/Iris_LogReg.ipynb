# AI ASSIGNMENT - LOGISTIC REGRESSION ON IRIS DATASET

```python
# Import packages
import pandas as pd 
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
#LOAD DATASET
iris_df=pd.read_csv("iris.csv",index_col='ID')
iris_df
```

```python
iris_df.info()
```

```python
#Check IF NA VALUES ARE PRESENT
iris_df.isna().sum()
```

```python
#Features selection and separation into X and Y variables
features=["Sepal.Length","Sepal.Width","Petal.Length","Petal.Width"]
target=['Species']

X=iris_df[features]
Y=iris_df[target]
```

# SPLIT THE DATASET

```python
#Split the data into 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

# TRAIN MODEL

```python
#CREATE AND FIT THE LOGISTIC MODEL
model = LogisticRegression()
model.fit(x_train, y_train) 
```

```python
#MAKE PREDICTIONS USING MODEL AND X_TEST 
predictions = model.predict(x_test)
predictions
```

# EVALUATE MODEL

```python
#Print Metrics: Precision, Recall and F1-score
print(classification_report(y_test, predictions))
accuracy_score=accuracy_score(y_test, predictions)
print(f"ACCURACY SCORE : {accuracy_score}")
```

```python
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt 
```

```python
confusion_matrix = metrics.confusion_matrix(y_test, predictions)
confusion_matrix
```

```python
#CONFUSION MATRIX USING SEABORN
plt.figure(figsize=(9,9))
sns.heatmap(confusion_matrix, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
all_sample_title = 'Accuracy Score: {0}'.format(accuracy_score)
plt.title(all_sample_title, size = 15)
```
