```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv(C:\Users\swati\OneDrive\Desktop\PG Diploma\Mall_Customers.csv")
```

```python
#Visual EDA __ Exploratory Data Analysis
sns.pairplot(data)
```

```python
data.info()
```

```python
#Feature Array
features = data.iloc[:,[3,4]].values
```

```python
#Implement Modelling

from sklearn.cluster import KMeans
model = KMeans(n_clusters=5)
model.fit(features)
```

```python
dataFinal = data.iloc[:,[3,4]]
dataFinal['label'] = model.predict(features) #Labels
dataFinal.head()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(dataFinal, hue="label", size= 7) \
.map(plt.scatter, "Annual Income (k$)", "Spending Score (1-100)") \
.add_legend();

plt.show()
```

```python
#Elbow Method to identify the ideal value of K(n_clusters) in KMeans algo
#Within Cluster Sum of Squares

features1 = data.iloc[:,[2,3,4]].values
from sklearn.cluster import KMeans
wcss=[]

for i in range(1,10):
    model = KMeans(n_clusters=i)
    model.fit(features1)
    wcss.append(model.inertia_)
    
plt.plot(range(1,10),wcss)
```
