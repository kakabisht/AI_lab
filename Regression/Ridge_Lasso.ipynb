# AI ASSIGNMENT - LINEAR , RIDGE and LASSO Regression on 50 Startup dataset

```python
#importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading and checking the dataset

```python
#read CSV
startup_df=pd.read_csv("50_Startups.csv")
```

```python
#Checking entries
startup_df.head(10)
```

```python
#Checking INFO of dataframe
startup_df.info()
```

```python
#Checking NA values
startup_df.isna().sum()
```

# Splitting and Scaling the dataset 

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# LABEL ENCODE THE STATE Column
label_encoder=LabelEncoder()
startup_df['STATE']= label_encoder.fit_transform(startup_df['STATE'])
startup_df.head(5)
```

```python
#SELECT FEATURES AND TARGET COLUMNS AND SEPARATE INTO X AND Y
features=["RND","ADMIN","MKT","STATE"]
target=['PROFIT']

X=startup_df[features]
Y=startup_df[target]
```

```python
#Test Train split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
```

```python
#SCALE X Train and X Test using STANDARD SCALER
scaler=StandardScaler()
scaled_X_train=X_train
scaled_X_train[features[:-1]]=scaler.fit_transform(scaled_X_train[features[:-1]])
scaled_X_train=scaled_X_train.values

scaled_X_test=X_test
scaled_X_test[features[:-1]]=scaler.transform(scaled_X_test[features[:-1]])
scaled_X_test=scaled_X_test.values
```

```python
#SCALE Y Train and Y Test using STANDARD SCALER
scaler=StandardScaler()

scaled_Y_train=scaler.fit_transform(Y_train)
scaled_Y_test=scaler.transform(Y_test)
```

# performing Multiple regression

```python
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm
```

```python
#CREATE AND FIT LINEAR REGRESSION MODEL
linear_model=LinearRegression()
linear_model.fit(scaled_X_train,scaled_Y_train)
```

```python
#MAKE PREDUCTIONS
predictions=linear_model.predict(scaled_X_test)
predictions
```

```python
#CALCULATE AND SHOW ERROR METRICS
linear_MSE=sm.mean_squared_error(scaled_Y_test,predictions)
print(f"Multiple Linear Regression Model -> Mean Squared Error(MSE) : {linear_MSE}")
RMSE=np.sqrt(linear_MSE)
print(f"Multiple Linear Regression Model -> RMSE : {RMSE}")
print("Multiple Linear Regression Model -> R2 score =", round(sm.r2_score(scaled_Y_test, predictions), 2))
print(f"Multiple Linear Regression Model -> Regression score : {linear_model.score(scaled_X_test,scaled_Y_test)}")
```

```python
plt.scatter(scaled_Y_test,predictions)
plt.plot([predictions.min(),predictions.max()],[predictions.min(),predictions.max()],"k--",lw=2,color="red")
```

# Performing Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
#CREATE AND FIT RIDGE REGRESSION MODEL
ridge_model=Ridge(alpha=0.001) 
ridge_model.fit(scaled_X_train,scaled_Y_train)
```

```python
#MAKE PREDICTIONS USING RIDGE REG. Model
ridge_predictions=ridge_model.predict(scaled_X_test)
ridge_predictions
```

```python
#CALCULATE ERROR METRICS FOR RIDGE REGRESSION MODEL
Ridge_train_score = ridge_model.score(scaled_X_train,scaled_Y_train)
print(f"Ridge regression model - Train score : {Ridge_train_score}")

ridge_MSE=sm.mean_squared_error(scaled_Y_test,ridge_predictions)
print(f"Ridge regression model - Mean Squared Error(MSE) : {ridge_MSE}")

ridge_RMSE=np.sqrt(ridge_MSE)
print(f"Ridge regression model - RMSE : {ridge_RMSE}")

print("Ridge regression model - R2 score =", round(sm.r2_score(scaled_Y_test, ridge_predictions), 2))
print(f"Ridge regression model - Regression score : {ridge_model.score(scaled_X_test,scaled_Y_test)}")
```

```python
plt.scatter(scaled_Y_test,ridge_predictions)
plt.plot([ridge_predictions.min(),ridge_predictions.max()],[ridge_predictions.min(),ridge_predictions.max()],"k--",lw=2,color="red")
```

# PERFORMING LASSO REGRESSION

```python
from sklearn.linear_model import Lasso
```

```python
#CREATE AND FIT LASSO REGRESSION MODEL
lasso_model= Lasso(alpha=0.001)
lasso_model.fit(scaled_X_train,scaled_Y_train)
```

```python
#MAKE PREDICTIONS USING LASSO REGRESSION MODEL
lasso_predictions=lasso_model.predict(scaled_X_test)
lasso_predictions
```

```python
#CALCULATE ERROR METRICS FOR LASSO REGRESSION MODEL
lasso_train_score = lasso_model.score(scaled_X_train,scaled_Y_train)
print(f"Lasso regression model - Train score : {lasso_train_score}")

lasso_MSE=sm.mean_squared_error(scaled_Y_test,lasso_predictions)
print(f"Lasso regression model - Mean Squared Error(MSE) : {ridge_MSE}")

lasso_RMSE=np.sqrt(lasso_MSE)
print(f"Lasso regression model - RMSE : {lasso_RMSE}")

print("Lasso regression model - R2 score =", round(sm.r2_score(scaled_Y_test, lasso_predictions), 3))
print(f"Lasso regression model - Regression score : {lasso_model.score(scaled_X_test,scaled_Y_test)}")
```

```python
plt.scatter(scaled_Y_test,lasso_predictions)
plt.plot([lasso_predictions.min(),lasso_predictions.max()],[lasso_predictions.min(),lasso_predictions.max()],"k--",lw=2,color="red")
```

# PERFORM ELASTICNET REGRESSION
## ElasticNet combines the properties of both Ridge and Lasso regression. It works by penalizing the model using both the L2 - Norm and the L1 - Norm.

```python
from sklearn.linear_model import ElasticNet
```

```python
#CREATE AND FIR ELASTICNET MODEL
elnt_model = ElasticNet(alpha = 0.001)
elnt_model.fit(scaled_X_train, scaled_Y_train) 
```

```python
#MAKE PREDICTIONS USING ElasticNet REGRESSION MODEL
elnt_predictions=elnt_model.predict(scaled_X_test)
elnt_predictions
```

```python
#CALCULATE ERROR METRICS FOR ElasticNet REGRESSION MODEL
elnt_train_score = elnt_model.score(scaled_X_train,scaled_Y_train)
print(f"ElasticNet regression model - Train score : {elnt_train_score}")

elnt_MSE=sm.mean_squared_error(scaled_Y_test,elnt_predictions)
print(f"ElasticNet regression model - Mean Squared Error(MSE) : {elnt_MSE}")

elnt_RMSE=np.sqrt(elnt_MSE)
print(f"ElasticNet regression model - RMSE : {elnt_RMSE}")

print("ElasticNet regression model - R2 score =", round(sm.r2_score(scaled_Y_test, elnt_predictions), 3))
print(f"ElasticNet regression model - Regression score : {elnt_model.score(scaled_X_test,scaled_Y_test)}")
```
